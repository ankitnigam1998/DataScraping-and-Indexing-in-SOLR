```python
sqlCommand = """
SELECT
        ROUND(Year(Release_date), -1) AS Decade,
        Round(Popularity, 2)          AS Popularity,
                                         Title,
                                         Artist
FROM df_table
INNER JOIN (SELECT Max(Popularity) as mp
            FROM df_table
            WHERE ROUND(Year(Release_date), -1) IS NOT NULL
            AND USA == 1
            GROUP BY ROUND(Year(Release_date), -1)
           ) AS temp
ON temp.mp = df_table.Popularity
ORDER BY Decade ASC, Popularity ASC
"""

output = pySpark.sql(sqlCommand)
output.toPandas().drop_duplicates(subset='Decade', keep="last")
```

```python
sqlCommand = """
SELECT
    ROUND(Year(Release_date), -1) as Decade,
    ROUND(Max(Popularity), 2) as Popularity,
    SUBSTRING(MAX(CONCAT(LPAD(Popularity, 11, 0), Title)), 12) AS Title,
    SUBSTRING(MAX(CONCAT(LPAD(Popularity, 11, 0), Artist)), 12) AS Artist
FROM
    df_table
WHERE
    ROUND(Year(Release_date), -1) IS NOT NULL
    AND USA == 1
GROUP BY Decade
ORDER BY Decade ASC
"""

pySpark.sql(sqlCommand).show()
```

```python
sqlCommand_kpop = """
SELECT Title, Artist, {}
FROM df_table
WHERE `k-pop` = 1
""".format(', '.join(supportingFeatures))

kpFrame = (pySpark.sql(sqlCommand_kpop)
                      .sample(.1)
                      .dropna()
                      .toPandas()
                )

sqlCommand_rap = """
SELECT Title, Artist, {}
FROM df_table
WHERE rap = 1
""".format(', '.join(supportingFeatures))

rapFrame = (pySpark.sql(sqlCommand_rap)
                     .sample(.1)
                     .dropna()
                     .toPandas()
               )
rapFrame.head()
```

```python
rapFrame = rapFrame.assign(is_rap = 1,
                                   is_kpop = 0
                                   )
kpFrame = kpFrame.assign(is_rap = 0,
                                     is_kpop = 1
                                     )
kpRapFrame = pd.concat([rapFrame, kpFrame])
X = scaler.fit_transform(kpRapFrame.iloc[:, 2:])

pca = PCA(n_components = 10)
pca.fit(X)
print(pca.explained_variance_ratio_)
```

```python
sns.lineplot(x=[x for x in range(1, 11)], y=pca.explained_variance_ratio_).set_title("% Variance Explained vs # Dimensions");
```

```python
pca = PCA(n_components=2)
pComps = pca.fit_transform(X)
pcaCompsFrame = pd.DataFrame(pComps)

sns.scatterplot(x = pcaCompsFrame[0], y = pcaCompsFrame[1], alpha = .2).set_title("First 2 PCA Components");
plt.xlabel('PCA 1');
plt.ylabel('PCA 2');
```

```python
clusters = range(1, 10)
inertias = []
for k in clusters:

    model = KMeans(n_clusters=k)

    model.fit(pcaCompsFrame.iloc[:,:2])

    inertias.append(model.inertia_)

sns.lineplot(x=clusters, y=inertias, marker='o').set_title("Inertia vs # Clusters used")
plt.xlabel('number of clusters, k')
plt.ylabel('inertia')
plt.xticks(clusters)
plt.show()
```

```python
km = KMeans(
    n_clusters=2, init='random',
    n_init=10, max_iter=300,
    tol=1e-04, random_state=0
)
y_km = km.fit_predict(X)
```

```python
clusterPlot = pd.concat([pcaCompsFrame, pd.Series(y_km)], axis = 1)
clusterPlot.columns = ['PCA_1', 'PCA_2', 'Cluster']
```

```python
sns.scatterplot(data=clusterPlot, x='PCA_1', y='PCA_2', hue='Cluster')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.title('2 component PCA');
```

```python
lastFrame = pd.concat([clusterPlot, kpRapFrame.reset_index()['is_rap']], axis=1)
lastFrame['is_rap'] = lastFrame['is_rap'].replace({1:'Rap', 0: 'KPop'})
lastFrame['Cluster'] = lastFrame['Cluster'].replace({1:'Cluster 2', 0: 'Cluster 1'})
lastFrame = lastFrame.rename(columns={'is_rap': 'Genre'})

sns.scatterplot(data=lastFrame, x='PCA_1', y='PCA_2', hue='Genre')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.title('2 component PCA');
```
